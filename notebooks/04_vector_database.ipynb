```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
import sys
from pathlib import Path

project_root = Path.cwd().parent
sys.path.append(str(project_root))
```

```python
# Load the PDF document

pdf_path = r"../data/raw/basel/bcbs189.pdf"

loader = PyPDFLoader(pdf_path)
documents = loader.load()
type(documents), len(documents)
```

```python
from src.text_splitter import split_documents

# Chunk the documents into smaller pieces

chunks = split_documents(documents)

type(chunks), len(chunks)
```

```python
# Generate embeddings for the chunks for questions
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# Build the FAISS vector store from the document chunks and their embeddings
vectorstore = FAISS.from_documents(
    documents=chunks,
    embedding=embedding_model
)

print(type(vectorstore))
```

```python
vectorstore.save_local("../vectorstore/faiss_index")
```
