### Module 1
#### Document Loading using LangChain
Learn how LangChain loads PDF documents into Document objects while preserving metadata.

```python
import sys
print(sys.executable)
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
pdf_path = r"../data/raw/basel/bcbs189.pdf"

loader = PyPDFLoader(pdf_path)
```

```python
documents = loader.load()
type(documents), len(documents)
```

```python
print(documents[0].page_content)  # Print the first page of the first document
```

```python
documents[0].metadata  # Print the metadata of the first document
```

```python
for page in documents[:3]:
    print("="*80)
    print(page.metadata)
    print(page.page_content[:500])
```
