# Module 2: Text Chunking

## Objective

Learn why chunking is required in Retrieval-Augmented Generation (RAG) systems and how LangChain splits documents into semantically meaningful chunks.

### Learning Outcomes

- Understand the limitations of page-based retrieval
- Learn the importance of chunk size and overlap
- Generate chunks using RecursiveCharacterTextSplitter
- Inspect generated chunks and metadata

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
pdf_path = r"../data/raw/basel/bcbs189.pdf"

loader = PyPDFLoader(pdf_path)
```

```python
documents = loader.load()
type(documents), len(documents)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)
```

```python
chunks = text_splitter.split_documents(documents)
```

```python
type(chunks), len(chunks)
```
