### Module 5 : Semantic Retrieval

#### Objective

Retrieve the most semantically relevant chunks from the FAISS vector database given a user query.

Learning Outcomes

- Understand similarity search
- Perform semantic retrieval
- Retrieve top-k documents
- Inspect retrieval quality

```python
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
vectorstore = FAISS.load_local(
    "../vectorstore/faiss_index",
    embedding_model,
    allow_dangerous_deserialization=True
)
```

```python
query = "What is LCR?"

results = vectorstore.similarity_search(
    query,
    k=3
)

len(results)
```

```python
for i, doc in enumerate(results, start=1):
    print("=" * 100)
    print(f"Retrieved Chunk {i}")
    print("=" * 100)
    print(doc.page_content)
    print()
```

```python
for doc in results:
    print(doc.metadata)
```

```python
queries = [
    "What is Liquidity Coverage Ratio?",
    "Explain leverage ratio",
    "What are capital buffers?",
    "What is counterparty credit risk?"
]

for q in queries:
    print("\n")
    print("="*120)
    print(q)
    print("="*120)

    docs = vectorstore.similarity_search(q, k=2)

    for d in docs:
        print(d.page_content[:100])
        print()
```
